```python
datafiles = glob.glob('Data/*.txt')
```

```python
data = {}
for f in datafiles:
        fdata = np.loadtxt(f)
        time = fdata[:,0]
        time -= time[0]
        power = fdata[:,1]
        temps = fdata[:,2]
        name = f.split('/')[-1].split('.')[0]
        data[name] = np.column_stack((time, power, temps))
```

```python
for key in data.keys():
    plot(data[key][:,0], data[key][:,1], label=key)
legend()
```
